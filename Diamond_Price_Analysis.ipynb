```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Make plots look better
sns.set(style="whitegrid", palette="muted", font_scale=1.1)
plt.rcParams['figure.figsize'] = (10, 6)

# Step 2: Load dataset
# Upload your diamonds.csv file manually via the upload option or mount Google Drive
from google.colab import files
uploaded = files.upload()

# Read the CSV
df = pd.read_csv(next(iter(uploaded)))

# Step 3: Clean & preprocess
df.drop(columns=['Unnamed: 0'], inplace=True, errors='ignore')
df.dropna(inplace=True)
df['size'] = df['x'] * df['y'] * df['z']  # Size = volume

# Step 4: Summary statistics
print("Summary Statistics:")
print(df[['carat', 'depth', 'table', 'price', 'size']].describe())
print(df.head())
```

```python
# Price Distribution
sns.histplot(df['price'], bins=50, kde=True, color='green')
plt.title("Diamond Price Distribution")
plt.xlabel("Price (USD)")
plt.ylabel("Count")
plt.show()
```

```python

# Carat vs Price (colored by Cut)
sns.scatterplot(data=df, x='carat', y='price', hue='cut', alpha=0.6)
plt.title("Carat vs Price by Cut")
plt.show()
```

```python

# Cut vs Price
sns.boxplot(data=df, x='cut', y='price')
plt.title("Price Distribution by Cut")
plt.show()
```

```python

# Clarity vs Price
sns.boxplot(data=df, x='clarity', y='price')
plt.title("Price Distribution by Clarity")
plt.show()
```

```python

# Color vs Price
sns.boxplot(data=df, x='color', y='price')
plt.title("Price Distribution by Color")
plt.show()
```

```python
# Size vs Price
sns.scatterplot(data=df, x='size', y='price', alpha=0.4)
plt.title("Size vs Price")
plt.xlabel("Volume (Size)")
plt.ylabel("Price")
plt.show()
```
